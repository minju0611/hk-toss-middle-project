## EDA

### 타겟 여부 (1, 0) 모두를 포함한 시각화

```python
#타겟 변수 분포 확인
plt.figure(figsize=(6, 4))
df["exited"].value_counts().plot(kind="bar", color=["skyblue", "salmon"])
plt.title("고객 이탈 여부 분포")
plt.xlabel("이탈 여부 (0: 이탈 없음, 1: 이탈)")
plt.ylabel("개수")
plt.xticks(rotation=0)
plt.show()
```

```python
# 고객 이탈 비율 확인
fig, ax = plt.subplots(figsize=(7,7))
count = Counter(df['exited'])
labels = ["잔류 고객", "이탈 고객"]
ax.pie(count.values(), labels = labels, autopct = lambda p: f'{p:.2f}%', colors=["skyblue", "salmon"])
ax.set_title('고객 이탈 비율')
plt.show()
```

전체 고객 중 **이탈 고객이 5.25%** 밖에 안되는 불균형 데이터

```python
# AGE를 연령대별로 변환하는 함수 정의
def categorize_age(age):
    if age in [20, 25]: return "20대"
    elif age in [30, 35]: return "30대"
    elif age in [40, 45]: return "40대"
    elif age in [50, 55]: return "50대"
    elif age in [60, 65]: return "60대"
    else: return "70대 이상"

# apply()를 사용하여 AGE_GROUP 생성
df["AGE_GROUP"] = df["AGE"].apply(categorize_age)

# 연령대별 이탈 여부 시각화
plt.figure(figsize=(8, 4))
sns.countplot(x=df["AGE_GROUP"], hue=df["exited"], palette=["skyblue", "salmon"], order=["20대", "30대", "40대", "50대", "60대", "70대 이상"])
plt.title("연령대별 이탈 여부")
plt.xlabel("연령대")
plt.ylabel("개수")
plt.show()
```

전체적으로 40대 고객이 많음

```python
df["SEX"] = df["SEX"].replace({1: "남자", 2: "여자"}).astype(str)

plt.figure(figsize=(6, 4))
sns.countplot(x=df["SEX"], hue=df["exited"], palette=["skyblue", "salmon"])
plt.title("성별 이탈 여부")
plt.xlabel("성별")
plt.ylabel("개수")
plt.show()
```

성별에는 큰 차이가 없어 보임(아주 미세하게 여자가 더 많음)

```python
# 직업군 별 한글 맵핑
job_mapping = {
    410: "대기업 급여소득",
    420: "일반 급여소득",
    430: "전문직 급여소득",
    440: "대표",
    510: "일반 자영업",
    520: "전문직 자영업",
    910: "기타/무직"
}
df["JB_TP_LABEL"] = df["JB_TP"].map(job_mapping)

# 직업군 별 이탈 고객 시각화
plt.figure(figsize=(6, 5))
sns.countplot(x=df["JB_TP_LABEL"], hue=df["exited"], palette=["skyblue", "salmon"], order = df["JB_TP_LABEL"].value_counts().index)
plt.title("이탈 여부에 따른 직업군 분포")
plt.xlabel("직업군")
plt.ylabel("개수")
plt.xticks(rotation=45)
plt.show()
```

일반 급여소득이 압도적으로 많음

```python
# HOME_ADM 코드 → 지역명 매핑
home_adm_mapping = {
    11: "서울", 26: "부산", 27: "대구", 28: "인천", 29: "광주", 30: "대전", 31: "울산", 36: "세종",
    41: "경기", 42: "강원", 43: "충북", 44: "충남", 45: "전북", 46: "전남", 47: "경북", 48: "경남", 50: "제주"
}
df["HOME_ADM_LABEL"] = df["HOME_ADM"].map(home_adm_mapping)

# HOME_ADM별 개수 기준으로 정렬된 순서 가져오기
sorted_order = df["HOME_ADM_LABEL"].value_counts().index

# 정렬된 순서대로 그래프 그리기
plt.figure(figsize=(8, 5))
sns.countplot(x=df["HOME_ADM_LABEL"], hue=df["exited"], palette=["skyblue", "salmon"], order=sorted_order)
plt.title("이탈 여부에 따른 거주 지역 분포")
plt.xlabel("지역")
plt.ylabel("개수")
plt.xticks(rotation=45)
plt.show()
```

경기, 서울이 압도적으로 많음

### 상관관계 분석

```python
# 필요한 열만 선택
df_selected = df[['SEX', 'AGE', 'JB_TP', 'HOME_ADM', 'exited']]

# 범주형 변수를 원-핫 인코딩
df_encoded = pd.get_dummies(df_selected, columns=['SEX', 'AGE', 'JB_TP', 'HOME_ADM'])

# 상관관계 계산
corr_matrix = df_encoded.corr()

# 'exited'과 관련된 상관관계만 추출
corr_exited = corr_matrix[['exited']].drop('exited')

# 히트맵 그리기
plt.figure(figsize=(8, 6))
sns.heatmap(corr_exited, annot=True, cmap='coolwarm', linewidths=0.5, fmt=".2f")
plt.title('범주형 변수와 Exited 간 상관관계 히트맵')
plt.show()
```
